```python
# Necessary for importing modules from a sub-directory
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
if os.getcwd().split('/')[-1] == 'notebooks':
    print("CHANGE DIR TO ROOT")
    os.chdir(r"../")
    
# Set mallet path
mallet_path ='mallet-2.0.8/bin/mallet'
    
# Import libraries
import pickle
from preprocess import build_ne_gold_phraser, get_gold_ngrams, apply_ngrams
import gensim
from gensim import matutils, models, corpora
```

```python
####### TRAIN MODEL
```

```python
# Preprocess variables

DATASET_TYPE = 'DN'
DATASET_SIZE = '10000'
DATASET_INDEX = '1'

# Load pickles:
data = pickle.load(open(f'pickles/efselab_parsed_{DATASET_TYPE}_{DATASET_SIZE}_{DATASET_INDEX}.pkl', 'rb'))

# Use n_grams or not
NGRAM = False # False | True
if NGRAM is True:
    NGRAM_OPT = 'gold+ne' # 'ext+int' | 'int' 
else:
    NGRAM_OPT = '-'
```

```python
if (NGRAM is True):
    # Load NE_grams and gold_grams
    NE_GRAMS_FILE = f'ngrams_per_dataset/ne_ngrams_{DATASET_TYPE}_{DATASET_SIZE}_{DATASET_INDEX}.pkl'
    ne_ngrams = pickle.load(open(NE_GRAMS_FILE, 'rb'))
    gold_ngrams = get_gold_ngrams()
    empty_grams = {
        'bigrams': [],
        'trigrams': []
    }

    if NGRAM_OPT == 'gold+ne':
        phraser = build_ne_gold_phraser(ne_ngrams, gold_ngrams)
    else:
        phraser = build_ne_gold_phraser(empty_grams, gold_ngrams)
        
    data = apply_ngrams(data, phraser)
```

```python
# Model variables

# Number of topics
NUM_TOPICS = 40

# Format Gensim components for data
id2word = corpora.Dictionary(data)
id2word.filter_extremes(no_below=2)
corpus = [id2word.doc2bow(text) for text in data]

# TRAIN MODEL
model = models.wrappers.LdaMallet(mallet_path,
                                  corpus=corpus,
                                  id2word=id2word,
                                  num_topics=NUM_TOPICS)
```

```python
####### CLASSIFICATION OF NEW ARTICLES
```

```python
# PARAMETERS FOR CLASSIFICATION DATA 

DATASET_TYPE = 'DN'
DATASET_SIZE = '10000'
DATASET_INDEX = '2'

CLASSIFICATION = True

if CLASSIFICATION:
    test_data = pickle.load(open(f'pickles/efselab_parsed_{DATASET_TYPE}_{DATASET_SIZE}_{DATASET_INDEX}.pkl', 'rb'))
    test_truncated = test_data
    id2word = corpora.Dictionary(test_truncated)
    other_corpus = [id2word.doc2bow(text) for text in test_truncated]
    topic_vector = model[other_corpus]
```
